```python
import numpy as np
a = np.array([21,4])
b =np.array([7,2]) 
a+b
np.subtract(a,b)
np.add(a,b)
np.multiply(a,b)
np.divide(a,b)
np.divmod(a,b)
np.append(a,b)
```

```python
import matplotlib.pyplot as plt
a = np.array([3,4])
b = np.array([7,2])
c = a+b

plt.quiver(a[0], a[1], angles='xy',scale_units='xy', scale=1)
plt.quiver(b[0], b[1], angles='xy',scale_units='xy', scale=1)
plt.quiver(c[0], c[1], angles='xy',scale_units='xy', scale=1, color='r');

plt.xlim(0,11)
plt.ylim(0,7)

plt.text(a[0], a[1], r'$\vec{A}$',size=15)
plt.text(b[0], b[1], r'$\vec{B}$',size=15)
plt.text(c[0], c[1], r'$\vec{C}$',size=15, c='r')
```

```python
a = np.array([3,4])
b = np.array([7,2])
d = a-b

plt.quiver(a[0], a[1], angles='xy',scale_units='xy', scale=1)
plt.quiver(b[0], b[1], angles='xy',scale_units='xy', scale=1)
plt.quiver(d[0], d[1], angles='xy',scale_units='xy', scale=1, color='y');

plt.xlim(-5,9)
plt.ylim(-1,5)
plt.axvline(x=0 , color= 'grey')
plt.axhline(y=0, color ='grey')

plt.text(a[0], a[1], r'$\vec{A}$',size=15)
plt.text(b[0], b[1], r'$\vec{B}$',size=15)
plt.text(d[0], d[1], r'$\vec{C}$',size=15, c='y')
```
